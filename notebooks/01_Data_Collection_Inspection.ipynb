## 1. Listings

```python
# loading listings first as its the main file
listings = pd.read_csv(listings_path, compression='gzip')

print("Shape:", listings.shape)
print("\nColumns:\n", listings.columns.tolist())
```

6,945 listings, 79 columns. Lot of columns — plenty of them are URLs and free text we won't need.

```python
listings.head()
```

- `price` is `'$86.00'` — string with dollar sign, needs converting
- `host_is_superhost` is `'t'`/`'f'` not boolean
- `neighbourhood_group_cleansed` looks completely empty
- some listings have NaN price already (rows 1, 4)

```python
# checking data types and how many nulls we have
listings.info()
```

3 columns are entirely null — `license`, `calendar_updated`, `neighbourhood_group_cleansed`. Dropping those.

`price` has 1,701 nulls — about 24% of listings. Can't do anything with those so they're getting dropped.

Review score nulls (1,042) are all the same listings — ones with zero reviews. That's fine, just means they're new.

```python
# need to check what format price is in before we do anything with it
print(listings['price'].dtype)
print("\nSample values:")
print(listings['price'].head(10))
```

Confirmed — `'$86.00'` format. Dollar sign even though it's Dublin data. Values are in euros, just stored with `$`.

```python
# how many missing values in each column
null_counts = listings.isnull().sum()
null_counts = null_counts[null_counts > 0].sort_values(ascending=False)

print("Columns with missing values:\n")
print(null_counts)
```

`license` is 100% null across all 6,945 rows — confirms what we noted in the scope, no compliance data available in this dataset.

Plan for cleaning:
- drop the 3 empty columns + URL/text cols
- drop rows with null price
- fill `bedrooms`, `beds`, `bathrooms` nulls with median
- leave host rate nulls as-is, don't want to fake that

## 4. GeoJSON

```python
import json

with open(geo_path, 'r') as f:
    geo = json.load(f)

print("Type:", geo['type'])
print("Number of neighbourhoods:", len(geo['features']))
print("\nAll neighbourhood names:")
for feature in geo['features']:
    print(" -", feature['properties'])
```

Only 4 boundaries — Dublin City, Dún Laoghaire-Rathdown, Fingal, South Dublin.

These are the 4 local authority areas, not granular neighbourhoods. Choropleth maps will be at this level.

```python
# neighbourhood_cleansed is our main geography column — check how many areas we have
print("Unique values in neighbourhood_cleansed:")
print(listings['neighbourhood_cleansed'].value_counts())
```

Matches the GeoJSON exactly — 4 areas. Dublin City has the bulk of listings as expected.

```python
# double checking the actual scrape date from the data itself
print("Scrape dates in dataset:")
print(listings['last_scraped'].value_counts())
```

All 6,945 listings scraped on the same date — clean single snapshot, no mixing of old and new scrapes.
